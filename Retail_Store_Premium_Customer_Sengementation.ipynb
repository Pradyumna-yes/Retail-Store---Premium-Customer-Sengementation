**Read the CSV file**

```python
import pandas as pd
df =pd.read_csv("/content/drive/MyDrive/DBS/Applied_Stats_Assigenment/YouTube Classes/Decision Tree/Sangeetha.csv")
#Display the first 5 rows of the DataFrame
df.head(5)
```

**Declare the target feature**

```python
inputs  = df.drop('Premium_Customer',axis='columns')
target = df['Premium_Customer']
```

**Perform label encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python

le_Customer =LabelEncoder()
le_Store = LabelEncoder()
le_Brand =LabelEncoder()
le_Model = LabelEncoder()
le_Price = LabelEncoder()
```

```python
inputs['le_Customer'] = le_Customer.fit_transform(inputs['Customer'])
inputs['le_Store'] = le_Store.fit_transform(inputs['Store'])
inputs['le_Brand'] = le_Brand.fit_transform(inputs['Brand'])
inputs['le_Model'] = le_Model.fit_transform(inputs['Model'])
inputs['le_Price'] = le_Price.fit_transform(inputs['Price'])
#Display the updated DataFrame
inputs.head()
```

**Display the updated DataFrame**

```python
#Drop the original categorical columns
inputs_n = inputs.drop(['Customer','Store','Brand','Model','Price'],axis='columns')
#Display the modified DataFrame
inputs_n
```

```python

from sklearn import tree
#Create a decision tree classifier
model = tree.DecisionTreeClassifier()
#Fit the model with the modified inputs and the target
model.fit (inputs_n,target)
#Calculate the accuracy score of the model
model.score(inputs_n,target)
```

```python
#Make predictions with the trained model
model.predict([[30,0,0,0,27]])
```
